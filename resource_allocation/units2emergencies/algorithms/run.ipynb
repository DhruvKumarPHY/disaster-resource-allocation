```python
import sys, os
sys.path.insert(1, '../')
import time, tqdm
import copy
import numpy as np
import pandas as pd
from operator import attrgetter
from utils import Emergency, EmergencyServiceLocation, Distribution, Input, Unit
from situations import situation1, situation2
from pso import PSO
from gwo import GWO
import statistics
import json, multiprocessing
```

```python
def get_results(dataset):
    f = open(file="dataset/"+dataset, mode="r")
    data = json.load(f)
    f.close()
    input = Input.from_situation(data)
    res = []
    for i in range(10):
        pso = PSO(input, iterations=1000, population=200, verbose=False, step=10, save=False)
        pso.run()
        res.append(pso.getBestFitness())
        
    pso_result = {"mean": statistics.mean(res), "std": statistics.stdev(res), "best": max(res), "worst": min(res)}
    
    res = []
    for i in range(10):
        gwo = GWO(input, iterations=1000, population=200, verbose=False, step=10, save=False)
        gwo.run()
        res.append(gwo.getBestFitness())
        
    gwo_result = {"mean": statistics.mean(res), "std": statistics.stdev(res), "best": max(res), "worst": min(res)}
    
    return {"pso": pso_result, "gwo": gwo_result}
```

```python
datasets = os.listdir("dataset/")
```

```python
a_pool = multiprocessing.Pool()
```

```python
results = a_pool.map(get_results, datasets[:1])
results
```
